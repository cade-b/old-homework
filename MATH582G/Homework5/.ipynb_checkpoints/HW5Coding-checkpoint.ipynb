```julia
using Plots, LinearAlgebra, ForwardDiff, Random, Statistics, Distributions, LaTeXStrings
```

```julia
function GD(x₀,f,∇f,β,maxiter)
    η = 1/β
    fvals = zeros(maxiter+1)
    x = x₀
    fvals[1] = f(x)
    for t = 1:maxiter
        x = x - η*∇f(x)
        #println(norm(∇f(x)))
        fvals[t+1] = f(x)
    end
    (x, fvals)
end
```

```julia
function AGD(x₀,f,∇f,β,maxiter)
    η = 1/β
    fvals = zeros(maxiter+1)
    x_old = x₀
    x = x₀
    a_old = 1.
    a = 1.
    fvals[1] = f(x)
    for t = 1:maxiter
        u = x + a*(1/a_old-1)*(x-x_old)
        x_old = x
        a_old = a
        x = u - η*∇f(x)
        #println(norm(∇f(x)))
        a = (√(a^4+4a^2)-a^2)/2
        fvals[t+1] = f(x)
    end
    (x, fvals)
end
```

## Question 2

```julia
n, m = 100, 80
λ = 1.;
```

```julia
Random.seed!(1234)
x_opt = randn(n)
A = randn(m,n)
ϵ = √0.25*randn(m)
y = A*x_opt+ϵ;
```

```julia
f(x) = norm(A*x-y,2)^2/2+(λ/2)*norm(x,2)^2;
```

```julia
∇f(x) = ForwardDiff.gradient(f,x);
```

```julia
derf(x) = A'*(A*x-y)+λ*x;
```

```julia
x₀ = randn(n)
β = 4(m+n)+λ
maxiter = 50;
```

```julia
(x, fvals) = GD(x₀,f,∇f,β,maxiter);
plot(0:maxiter,fvals,yaxis=:log,label=:false)
xlabel!(L"t")
ylabel!(L"f(x_t)")
title!("GD")
```

```julia
savefig("prob2a.pdf")
```

```julia
(x, fvals) = AGD(x₀,f,∇f,β,maxiter);
plot(0:maxiter,fvals,yaxis=:log,label=:false)
xlabel!(L"t")
ylabel!(L"f(x_t)")
title!("AGD")
```

```julia
savefig("prob2b.pdf")
```

## Question 3

```julia
n, m = 100, 80
λ = 1.;
```

```julia
Random.seed!(123)
x_opt = randn(n)
A = randn(m,n)
ϵ = √0.25*randn(m)
ind = rand(1:m,5)
println(ind) #make sure these are all distinct
ϵ[ind] = 5*randn(5)
y = A*x_opt+ϵ;
```

```julia
η = 5.;
```

```julia
h_η(w) = abs(w)≤η ? w^2/2 : η*(abs(w)-η/2);
```

```julia
f(x) = sum(h_η.(y-A*x))
∇f(x) = ForwardDiff.gradient(f,x);
```

```julia
x₀ = randn(n)
β = 2(m*η+λ)
maxiter = 100;
```

```julia
(x, fvals) = GD(x₀,f,∇f,β,maxiter);
plot(0:maxiter,fvals,yaxis=:log,label=:false)
xlabel!(L"t")
ylabel!(L"f(x_t)")
title!("GD")
```

```julia
savefig("prob3a.pdf")
```

```julia
(x, fvals) = AGD(x₀,f,∇f,β,maxiter);
plot(0:maxiter,fvals,yaxis=:log,label=:false)
xlabel!(L"t")
ylabel!(L"f(x_t)")
title!("AGD")
```

```julia
savefig("prob3b.pdf")
```
