```julia
using Plots, LaTeXStrings
```

```julia
p₀(x) = (-2<=x<=-1 || 1<=x<=2 || 6<=x<=7)
```

```julia
function q(x,t)
    0.5*[p₀(x-c₀*t)+p₀(x+c₀*t); (p₀(x-c₀*t)-p₀(x+c₀*t))/Z₀]
end
```

```julia
c₀, Z₀ = 1.0, 0.7
```

```julia
xvec = 0:0.01:4
tvec = [0. 0.5 1. 1.5 2. 3.]
```

```julia
for (j,t) in enumerate(tvec)
    p1 = plot(xvec,map(x -> q(x,t)[1],xvec),label=L"p(x,t)")
    plot!(xvec,map(x -> q(x,t)[2],xvec),label=L"u(x,t)")
    xlabel!(L"x")
    title!(L"t=%$t")
    display(p1)
    savefig(p1,"prob3-5-" * string(j) * ".pdf")
end
```

```julia
q(1.51,0.5)
```
